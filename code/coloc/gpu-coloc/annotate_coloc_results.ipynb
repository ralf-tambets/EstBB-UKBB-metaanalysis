```python
import pandas as pd
import os
import numpy as np

pop_dir_tag = {
    "EstBB":"Est","UKBB_AFR":"AFR","UKBB_AMR":"AMR","UKBB_CSA":"CSA",
    "UKBB_EAS":"EAS","UKBB_EUR":"EUR","UKBB_MID":"MID", "meta_EUR":"meta_EUR"
}

os.makedirs("annotated_gpu-coloc_results/FinnGen+MVP+UKBB_coloc", exist_ok=True)

for key, value in pop_dir_tag.items():
    metadata = pd.read_csv(f"annotated_meta_summaries/{key}_summary_annotated.tsv", sep="\t")

    metadata = metadata[["signal", "metabolite", "maf", "log10p", "nearest_genes", 'gene_symbols']].drop_duplicates()

    metadata = metadata.rename(columns={"signal":"signal2", "maf": "maf2", "log10p": "log10p2"})

    df = pd.read_csv(f"meta_colocid/FinnGen+MVP+UKBB_coloc/{key}_FinnGen+MVP+UKBB_META_results.tsv", sep="\t")

    meta_metadata = pd.read_csv(f"FinnGen+MVP+UKBB_important_variants.tsv", sep="\t")

    meta_metadata["log10p1"] = -np.log10(meta_metadata["all_inv_var_meta_p"])

    meta_metadata["maf1_FinnGen"] = meta_metadata["fg_af_alt"].apply(lambda x: min(x, 1-x) if pd.notna(x) else None)
    meta_metadata["maf1_MVP_EUR"] = meta_metadata["MVP_EUR_af_alt"].apply(lambda x: min(x, 1-x) if pd.notna(x) else None)

    meta_metadata = meta_metadata.rename(columns={"variant": "lead1"})
    
    meta_metadata = meta_metadata[["log10p1", "maf1_FinnGen", "maf1_MVP_EUR", "trait", "lead1"]]
    
    df = df[df["PP.H4"]>=0.8]

    df = df.merge(metadata, on=("signal2"), how="left")

    df["trait"] = (
        df["signal1"]
        .str.replace(r"_chr.*", "", regex=True)
        .str.removeprefix("FinnGen+MVP+UKBB_")
    )

    df = df.merge(meta_metadata, on=("trait", "lead1"), how="left")

    df.drop_duplicates(inplace=True)

    df.to_csv(f"annotated_gpu-coloc_results/FinnGen+MVP+UKBB_coloc/{value}_FinnGen+MVP+UKBB_coloc_results.tsv", sep="\t", index=False)
```

```python
import pandas as pd
import os

pop_dir_tag = {
   "UKBB_AFR":"AFR","UKBB_AMR":"AMR","UKBB_CSA":"CSA",
    "UKBB_EAS":"EAS","UKBB_EUR":"EUR","UKBB_MID":"MID", "meta_EUR":"meta_EUR",
    "EstBB": "EstBB"
}

os.makedirs("annotated_coloc_results/eQTL_Catalogue_coloc", exist_ok=True)

ensemble = pd.read_csv("ensemble.tsv", sep="\t")

eQTLC_metadata = pd.read_csv("tabix_ftp_paths.tsv", sep="\t")

eQTLC_metadata = eQTLC_metadata.rename(columns={
    'dataset_id': 'eqtl_catalogue_dataset_id',
    'quant_method': 'eqtl_catalogue_quant_method'
})

dataset_to_quant = dict(zip(eQTLC_metadata["eqtl_catalogue_dataset_id"], eQTLC_metadata["eqtl_catalogue_quant_method"]))

ensemble = ensemble.rename(columns={
    'molecular_trait_id': "eqtl_catalogue_molecular_trait_id",
    'gene_id': 'eqtl_catalogue_gene_id',
    'gene_name': 'eqtl_catalogue_gene_name'
})

for key, value in pop_dir_tag.items():
    for method in ["exon", "ge_microarray_aptamer", "leafcutter", "tx", "txrev"]:
        try:

            metadata = pd.read_csv(f"annotated_meta_summaries/{key}_summary_annotated.tsv", sep="\t")

            metadata = metadata[["signal", "metabolite", "maf", "log10p", "nearest_genes", "gene_symbols"]].drop_duplicates()

            metadata = metadata.rename(columns={"signal":"signal2", "maf": "maf2", "log10p": "log10p2"})

            df = pd.read_csv(f"meta_colocid/eQTL_catalogue_results/{method}_{key}_results.tsv", sep="\t")

            df = df[df["PP.H4"]>=0.8]

            df = df.merge(metadata, on=("signal2"), how="left")

            print(df.shape[0])

            df["eqtl_catalogue_quant_method"] = df["signal1"].str.split("_").str[0].map(dataset_to_quant)
            df["eqtl_catalogue_molecular_trait_id"] = df["signal1"].str.extract(r"^QTD[^_]*_(.*?)_L")

            df = df.merge(ensemble, how="left", on="eqtl_catalogue_molecular_trait_id")

            print(df.shape[0])

            df.drop_duplicates(inplace=True)

            df.to_csv(f"annotated_coloc_results/eQTL_Catalogue_coloc/{method}_{key}_results.tsv", sep="\t", index=False)    
        except:
            continue
```

```python
import pandas as pd
import os
from tqdm import tqdm

pop_pairs = [
            ["meta_EUR", "AFR"],["meta_EUR", "EUR"],["meta_EUR", "AMR"],["meta_EUR", "META"],
            ["UKBB_EUR", "EUR"],["EstBB", "EUR"],["UKBB_AFR", "AFR"],["UKBB_AMR", "AMR"],
            ["UKBB_EUR", "META"],["EstBB", "META"],["UKBB_AFR", "META"],["UKBB_AMR", "META"],["UKBB_CSA", "META"],["UKBB_EAS", "META"],
            ]

os.makedirs("annotated_gpu-coloc_results/MVP_coloc", exist_ok=True)

for eur, mvp in tqdm(pop_pairs):
    metadata = pd.read_csv(f"annotated_meta_summaries/{eur}_summary_annotated.tsv", sep="\t")

    metadata = metadata[["signal", "metabolite", "maf", "log10p", "nearest_genes", "gene_symbols"]].drop_duplicates()

    metadata = metadata.rename(columns={"signal":"signal2", "maf": "maf2", "log10p": "log10p2"})

    df = pd.read_csv(f"MVP_coloc/MVP_{mvp}_{eur}_results.tsv", sep="\t")

    df = df[df["PP.H4"]>=0.8]

    df = df.merge(metadata, on=("signal2"), how="left")

    df["MVP_trait_id"] = df["signal1"].str.split(".").str[2].str.split("_chr").str[0]
    df["MVP_pop"] = df["signal1"].str.split("_").str[0]

    MVP_traits_df = pd.read_csv(f"annotated_MVP_summaries/{mvp}_summary.tsv", sep="\t")

    print(df.shape[0])

    MVP_traits_df = MVP_traits_df.rename(columns={"lead_variant1": "lead1"})

    df = df.merge(MVP_traits_df, on=("signal1","MVP_trait_id", "MVP_pop", "lead1"))

    print(df.shape[0])

    df.drop_duplicates(inplace=True)

    df.to_csv(f"annotated_gpu-coloc_results/MVP_coloc/MVP_{mvp}_{eur}_results.tsv", sep="\t", index=False)
```
